limpieza de flujo de datos de los datos de alexa

 Definimos un pipeline para limpiar todos los mensajes
 
 El pipeline realizamos lo siguiente: (1) eliminar signos de putuación, (2) eliminar las stopword

```python
def message_cleaning(message):
    Test_punc_removed = [char for char in message if char not in string.punctuation]
    Test_punc_removed_join = ''.join(Test_punc_removed)
    Test_punc_removed_join_clean = [word for word in Test_punc_removed_join.split() if word.lower() not in stopwords.words('english')]
    return Test_punc_removed_join_clean
```

limpiamos toda la data set

```python
reviews_df_clean=reviews_df['verified_reviews'].apply(message_cleaning)
```

```python
print(reviews_df['verified_reviews'][3])
print(reviews_df_clean[3])
```

```python

vectorizer=CountVectorizer(analyzer=message_cleaning)
reviews_contvectorizer=vectorizer.fit_transform(reviews_df['verified_reviews'])
```

variable independiente

```python
X=reviews_df.drop(['feedback'],axis=1)
X
```

variable a predecir

```python
y=reviews_df['feedback']
y
```

modelos de clasificacion

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test, y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.naive_bayes import MultinomialNB

nb_multinomial=MultinomialNB()
nb_multinomial.fit(X_train,y_train)


```

eficacia

```python
from sklearn.metrics import classification_report, confusion_matrix,accuracy_score
```

```python
y_predict_train=nb_multinomial.predict(X_train)
y_predict_train
```

prueba de  entrenamiento

```python
cm=confusion_matrix(y_train,y_predict_train)

sns.heatmap(cm, annot=True)
```

prueba de test

```python
y_test_predict=nb_multinomial.predict(X_test)

cm2=confusion_matrix(y_test,y_test_predict)

sns.heatmap(cm2, annot=True)
```

```python
print(classification_report(y_test,y_test_predict))
```

Regresion logistica


```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
model.fit(X_train,y_train)

y_pred=model.predict(X_test)
```

```python
y_pred
```

```python
print("Accuracy {} %".format(100*accuracy_score(y_pred,y_test)))
```

```python
cm3=confusion_matrix(y_test,y_pred)
sns.heatmap(cm3,annot=True)
```

```python
print(classification_report(y_test,y_pred))
```
